```python
import csv
import os
import msvcrt as m


def tecla_para_continuar():
    print("\n\nPressione qualquer tecla para continuar")
    m.getch()

def ler_ficheiro_csv(caminho_ficheiro):
    dados = []
    with open(caminho_ficheiro, 'r') as ficheiro:
        leitor = csv.DictReader(ficheiro)
        next(leitor)
        for linha in leitor:
            dados.append(linha)
    return dados


def distribuicao_por_genero(dados):
    distribuicao_genero = {'Male': {'0': 0, '1': 0}, 'Female': {'0': 0, '1': 0}}
    for entrada in dados:
        genero = entrada['gender']
        diabetes = entrada['diabetes']
        if genero in distribuicao_genero:
            distribuicao_genero[genero][diabetes] += 1
        else:
            pass
    return distribuicao_genero


def distribuicao_por_idade(dados):
    distribuicao_idade = {'0-10': {'0': 0, '1': 0}, '11-24': {'0': 0, '1': 0}, '25-29': {'0': 0, '1': 0},
                       '30-34': {'0': 0, '1': 0}, '35-39': {'0': 0, '1': 0}, '40-44': {'0': 0, '1': 0}, '45+': {'0': 0, '1': 0}}
    for entrada in dados:
        idade = float(entrada['age'])
        diabetes = entrada['diabetes']
        if idade <= 10:
            grupo_idade = '0-10'
        elif idade <= 24:
            grupo_idade = '11-24'
        elif idade <= 29:
            grupo_idade = '25-29'
        elif idade <= 34:
            grupo_idade = '30-34'
        elif idade <= 39:
            grupo_idade = '35-39'
        elif idade <= 44:
            grupo_idade = '40-44'
        else:
            grupo_idade = '45+'
        distribuicao_idade[grupo_idade][diabetes] += 1
    return distribuicao_idade


def distribuicao_por_glucose(dados):
    distribuicao_glucose = {}
    for entrada in dados:
        glucose = int(entrada['blood_glucose_level'])
        intervalo_glucose = (glucose // 10) * 10
        if intervalo_glucose not in distribuicao_glucose:
            distribuicao_glucose[intervalo_glucose] = {'0': 0, '1': 0}
        diabetes = entrada['diabetes']
        distribuicao_glucose[intervalo_glucose][diabetes] += 1
    return distribuicao_glucose


def imprimir_tabela_distribuicao(distribuicao):
    print("Intervalo de Glucose\tSem Diabetes\tDiabetes")
    for chave in sorted(distribuicao.keys()):
        valores = distribuicao[chave]
        print(f"{chave}-{chave+9}\t{valores['0']}\t{valores['1']}")


caminho_ficheiro = r"C:\Users\rf\Downloads\diabetes_prediction_dataset.csv"
dados = ler_ficheiro_csv(caminho_ficheiro)


escolha = "1"
while (escolha != "0"):
    os.system('cls')
    print("\nMenu:\n")
    print("1 -> Distribuição da doença por género")
    print("2 -> Distribuição da doença por grupos etários")
    print("3 -> Distribuição da doença por níveis de glucose")
    print("0 -> Sair")
    
    escolha = input("\nEscolha a sua opção: ")
    os.system('cls')
    
    if escolha == '1':
        print("Distribuição da doença por género:")
        distribuicao_genero = distribuicao_por_genero(dados)
        print("Género\tSem Diabetes\tDiabetes")
        for genero, valores in distribuicao_genero.items():
            print(f"{genero}\t{valores['0']}\t{valores['1']}")
    elif escolha == '2':
        print("Distribuição da doença por grupos etários:")
        distribuicao_idade = distribuicao_por_idade(dados)
        print("Grupo Etário\tSem Diabetes\tDiabetes")
        for grupo_idade, valores in distribuicao_idade.items():
            print(f"{grupo_idade}\t{valores['0']}\t{valores['1']}")
    elif escolha == '3':
        print("Distribuição da doença por níveis de glucose:")
        distribuicao_glucose = distribuicao_por_glucose(dados)
        imprimir_tabela_distribuicao(distribuicao_glucose)
    elif escolha == '0':
        print("A sair do programa.")
    else:
        print("Opção inválida. Por favor, introduza uma opção válida.")
    tecla_para_continuar()
```
