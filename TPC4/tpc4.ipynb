```python
import os
import random
import msvcrt as m


def tecla_para_continuar():
    print("\n\nPressione qualquer tecla para continuar")
    m.getch()

def criar_lista():
    lista = [random.randint(1, 100) for _ in range(5)]
    print("Lista criada com sucesso")
    return lista

def reescrever_lista():
    global lista
    lista = criar_lista()

def ler_lista(lista):
    os.system('cls')
    print("Lista atual:", lista)

def soma(lista):
    return sum(lista)

def media(lista):
    return sum(lista) / len(lista)

def maior(lista):
    return max(lista)

def menor(lista):
    return min(lista)

def esta_ordenada_crescente(lista):
    return lista == sorted(lista)

def organizar_crescente(lista):
    lista = lista.sort()
    print("\nLista organizada com sucesso")
    return lista

def organizar_decrescente(lista):
    lista = lista.sort(reverse=True)
    print("\nLista organizada com sucesso")
    return lista

def esta_ordenada_decrescente(lista):
    return lista == sorted(lista, reverse=True)

def procurar_elemento(lista, elemento):
    return elemento in lista


lista = []

while True:
    os.system('cls')
    print("Opções do menu:\n\n")
    print("1 -> Criar lista")
    print("2 -> Ler lista")
    print("3 -> Soma")
    print("4 -> Média")
    print("5 -> Maior")
    print("6 -> Menor")
    print("7 -> Está ordenada por ordem crescente")
    print("8 -> Está ordenada por ordem decrescente")
    print("9 -> Procurar um elemento")
    print("0 -> Sair\n")

    opcao = int(input("Escolha uma opção: "))

    if opcao == 1:
        os.system('cls')
        reescrever_lista()
        tecla_para_continuar()
    elif opcao == 2:
        os.system('cls')
        ler_lista(lista)
        tecla_para_continuar()
    elif opcao == 3:
        os.system('cls')
        print("Soma:", soma(lista))
        tecla_para_continuar()
    elif opcao == 4:
        os.system('cls')
        print("Média:", media(lista))
        tecla_para_continuar()
    elif opcao == 5:
        os.system('cls')
        print("Maior:", maior(lista))
        tecla_para_continuar()
    elif opcao == 6:
        os.system('cls')
        print("Menor:", menor(lista))
        tecla_para_continuar()
    elif opcao == 7:
        os.system('cls')
        if esta_ordenada_crescente(lista):
            print("A lista está ordenada em ordem crescente.")
        else:
            organizar = input("A lista não está ordenada em ordem crescente.\nQuer organizar a lista?  [sim/nao]")
            if (organizar == "sim" or organizar == "s"):
                organizar_crescente(lista)
            else:
                os.system('cls')
        tecla_para_continuar()
    elif opcao == 8:
        os.system('cls')
        if esta_ordenada_decrescente(lista):
            print("A lista está ordenada em ordem decrescente.")
        else:
            organizar = input("A lista não está ordenada em ordem decrescente.\nQuer organizar a lista?  [sim/nao]")
            if (organizar == "sim" or organizar == "s"):
                organizar_decrescente(lista)
            else:
                os.system('cls')
        tecla_para_continuar()
    elif opcao == 9:
        os.system('cls')
        elemento = int(input("Digite o número a procurar na lista: "))
        if procurar_elemento(lista, elemento):
            print(f"\nO elemento {elemento} está na lista.")
        else:
            print(f"\nO elemento {elemento} não está na lista.")
        tecla_para_continuar()
    elif opcao == 0:
        os.system('cls')
        print("Obrigado por ter usado o programa ;)")
        m.getch()
        break
    else:
        os.system('cls')
        print("Opção inválida. Tente novamente.")
        tecla_para_continuar()
```
